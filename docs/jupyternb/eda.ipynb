# `EDA` class
- This class is a platform for the EDA analysis including:
- PCA
    - data visualization with a scatter plot
    - visualization of PC components (an alias for eigenvectors used in `skelearn.decomposition.PCA`)
- Definition of upregulated/downregulated genes by Cohen's d values
    - visualization with scatter plots (we name them "butterfly plots")
    - visualization with Venn diagrams

```python
from tools import SuematsuData, EDA, eda_longitudinal_args
```

## Documentation
### 1. `__init__`
- arguments
    - `data`: [`SuematsuData`](https://takemura-hgf.readthedocs.io/en/latest/jupyternb/suematsudata.html)
    - `out`: str (default:"/home/jovyan/out")
        - dirname for exporting output files
    - `adgile`: bool (default:False)
        - argument for `pipeline` function
        - pass True to skip analyses for existing output files
    - `args`: dict (defaut:{})
        - pass `eda_longitudinal_args` for longitudinal comparison
- instance variables
    - `out`: str
    - `args`: dict
        - these variables corresponds to respective arguments
    - `data`: `SuematsuData`
        - optional variable generated if there are more than one functions to be executed (taken account when `adgile=True`)
        - corresponds to `data` argument
    - `model` `sklearn.decomposition.PCA`
    - `pca`: `pandas.DataFrame`
        - optional variables generated if there are more than one functions to be executed (taken account when `adgile=True`)
- return: None
- output: N/A

```python
eda1 = EDA(
    data=SuematsuData(),
    args=eda_longitudinal_args
)

print(f"""
eda1.out: {eda1.out}\n
eda1.data: {eda1.data}\n
eda1.model: {eda1.model}\n
""")
```

### 2. `scatter_plot`
- arguments
    - `figsize`: tuple (default:(3, 3))
    - `**kwargs`
        - kwargs for `seaborn.scatterplot`
- return: None
- output: generates `pca.png` in the designated directory

```python
eda1.scatter_plot()
```

### 3. `component_plot`
- arguments
    - `layout`: tuple (default:(1, 2))
    - `figsize`: tuple (default:(10, 3))
        - arguments for `matplotlib.pyplot.subplots`
        - equivalent to `matplotlib.pyplot.subplots(*layout, figsize=figsize)`
    - `top`: int (default:30)
        - number of genes to show
        - double the number you want to show for both genes in top & bottom when `showbottom=True` (e.g., pass `top=30` to show top 15 genes and bottom 15 genes)
    - `show_bottom`: bool (default:True)
        - pass False to show only genes in top n (where n indicates the number you pass for `top`)
    - `**kwargs`
        kwargs for `seaborn.scatterplot`
- return: `None`
- output: generates `components.png` in the designated directory

```python
eda1.component_plot()
```

### 4. `butterfly_plot`
- arguments
    - `set_labels`: tuple (default:("day2", "day7"))
        - labels of comparison targets
        - **Note**: this argument is automatically adjusted when you pass `args=eda_longitudinal_args` in `__init__`
    - `layout`: tuple (default:(1, 2))
    - `figsize`: tuple (default:(10, 3))
        - arguments for `matplotlib.pyplot.subplots`
        - equivalent to `matplotlib.pyplot.subplots(*layout, figsize=figsize)`
    - `hspace`: float (default:0.4)
        - argument for `matplotlib.pyplot.Figure.subplots_adjust`
    - `d`: float (default:0.8)
        - Cohen's d value for the threshold.
        - upregulated genes will be defined to show larger Cohen's d values than the threshold (>`d`)
        - downregulated genes will be defined to show smaller Cohen's d values than the threshold (<`d`)
    - `flip`: bool (default:False)
        - pass True to flip the definition of upregulated/downregulated genes
    - `dim_idx`: int (default:1)
    - `idx_starts_with`: int (default:1)
        - arguments for designating PCs to choose as the y-axis
        - `dim_idx` is the pythonic index of the PC (e.g., `dim_idx=0` for the first PC)
        - `idx_starts_with` is the index of the PC with the largest contribution rate (e.g., `idx_starts_with=0` when you rather call it PC0)
        - **Note**: this argument is automatically adjusted when you pass `args=eda_longitudinal_args` in `__init__`
    - `palette`: str (default:"coolwarm")
    - `s`: float (default:5)
    - `alpha`: float (default:0.1)
        - arguments for `seaborn.scatterplot`
- return: `None`
- output: generates output img files in the designated directory
    - `butterfly_plot_day2_day7.png` will be generated when `set_labels=("day2", "day7")`
    - `butterfly_plot_HGF+_control.png` will be generated when you pass `args=eda_longitudinal_args` in `__init__`

```python
eda1.butterfly_plot()
```

### 5. `gene_regulation_venn_diagram`
- arguments
    - `layout`: tuple (default:(1, 2))
    - `figsize`: tuple (default:(10, 3))
        - arguments for `matplotlib.pyplot.subplots`
        - equivalent to `matplotlib.pyplot.subplots(*layout, figsize=figsize)`
    - `set_labels`: tuple (default:("day2", "day7"))
        - labels of comparison targets
        - **Note**: this argument is automatically adjusted when you pass `args=eda_longitudinal_args` in `__init__`
    - `set_colors`: tuple (default:("y", "m", "grey"))
        - colors in the venn diagram
        - argument for [`matplotlib_venn.venn2`](https://pypi.org/project/matplotlib-venn/)
        - **Note**: this argument is automatically adjusted when you pass `args=eda_longitudinal_args` in `__init__`
- return: `None`
- output: generates output img files in the designated directory
    - `venn_day2_day7.png` will be generated when `set_labels=("day2", "day7")`
    - `venn_HGF+_control.png` will be generated when you pass `args=eda_longitudinal_args` in `__init__`

```python
eda1.gene_regulation_venn_diagram()
```

### 6. `pipeline`
- arguments
    - pipe: list (default:`eda_pipeline_adgile.keys()`)
        - name of functins to call in the pipeline
        - **Note**: for speeding up `metapipeline` func, `scatter_plot` and `component_plot` are omitted when you pass `args=eda_longitudinal_args` in `__init__`
    - close: bool (default:False)
        - pass True to close generated imgs
    - adgile: bool (default:False)
        - pass True to skip analyses for existing output files
- return: `None`
- output: generates output img files in the designated directory

```python
%time
eda1.pipeline(close=False)
```

```python
%time
eda1.pipeline(close=True, adgile=True)
```
